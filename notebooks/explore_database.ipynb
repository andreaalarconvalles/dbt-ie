````python
import duckdb

# Connect to the database (relative path from notebooks/ directory)
con = duckdb.connect("../my_database.duckdb")

# we will use .pl() to display query results as polars dataframes
````

````python
# Show all tables in the main schema
con.sql("SHOW TABLES").pl()
````

````python
# Describe a specific table (replace 'table_name' with a real table from the list above)
con.sql("DESCRIBE customers").pl()
````

````python
# Sample data from a table
con.sql("SELECT * FROM customers LIMIT 10").pl()
````

````python
# Query information_schema to see columns information for a specific table
# Using .pl() to convert to a Polars DataFrame allows you to scroll and see all columns
con.sql("SELECT * FROM information_schema.columns WHERE table_name = 'customers'").pl()
````

````python
# Close the connection to the database
con.close()
````
